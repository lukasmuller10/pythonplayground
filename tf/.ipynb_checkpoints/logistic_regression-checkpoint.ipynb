```python
def convert(param):
    if param == ' >50K':
        return 0
    else:
        return 1
```

```python
Df.income = Df.income.apply(convert)
```

```python
Df.head()
```

```python
X = Df.drop('income',axis = 1)
X.head()
```

```python
Y = Df.income
Y
```

```python
Df.age.hist()
```

```python
import tensorflow as tf
```

```python
idade = tf.feature_column.numeric_column('age')
```

```python
idade_categorica = [tf.feature_column.bucketized_column(idade, boundaries = [20,30,40,50,60,70,80,90])]
```

```python
X.columns
```

```python
nome_colunas_categoricas = ['workclass','education','marital-status', 'occupation', 'relationship', 'race', 'sex',
            'native-country']
```

```python
colunas_categoricas = [tf.feature_column.categorical_column_with_vocabulary_list(key=c, vocabulary_list=X[c].unique()) for c in nome_colunas_categoricas]
```

```python
colunas_numericas = ['final-weight', 'education-num','capital-gain', 'capital-loos']
```

```python
colunas_numericas = [tf.feature_column.numeric_column(key = c) for c in colunas_numericas]
```

```python
colunas = idade_categorica + colunas_categoricas + colunas_numericas
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.3)
```

```python
training_func = tf.estimator.inputs.pandas_input_fn(x = X_train, y= Y_train,
                                                   batch_size=32, num_epochs=None, shuffle=True)
```

```python
classificador = tf.estimator.LinearClassifier(feature_columns=colunas)
```

```python
classificador.train(input_fn = training_func, steps=10000)
```

```python
func_previsao = tf.estimator.inputs.pandas_input_fn(x=X_test, y=Y_test,batch_size=32,shuffle=False)
```

```python
previsao = classificador.predict(input_fn = func_previsao)
```

```python
previsao_lista = []
for p in classificador.predict(input_fn=func_previsao):
    previsao_lista.append(p['class_ids'])
```

```python
previsao_lista
```

```python
taxa = accuracy_score(Y_test, previsao_lista)
taxa
```
